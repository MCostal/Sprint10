# CÁLCULOS ESTADÍSTICOS PRIMERA APROXIMACION

Se realiza la prueba de normalidad de Shapiro-Wilk para comprobar si nuestras variables siguen una distribución gaussiana. 
Esto nos marcará que tipo de análisis estadístico debemos aplicar posteriormente: parámetrico o no parámetrico.
Si siguen una distribución gaussiana, se aplicarán métodos estadísticos parámetricos, como Test de Pearson
Si no es una distribución gaussiana, aplicaremos métodos estadísticos no parámetricos, como el test de Spearman.

### 1. Prueba de Shapiro-Wilk


Comprueba si una muestra de datos tiene una distribución gaussiana.
Suposiciones: Las observaciones en cada muestra son independientes y se distribuyen de manera idéntica.
Interpretación:
H0: la muestra tiene una distribución gaussiana.
H1: la muestra no tiene una distribución gaussiana.

Hacemos el test de Normalidad para poder escoger después que método estadístico utilizaremos en la matriz de correlación.

```python
from scipy.stats import shapiro
```

Variable concnetración URM 
# Example of the Shapiro-Wilk Normality Test

```python
data = totales_finales['concentracion_URM']
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probably Gaussian')
else:
 print('Probably not Gaussian')
```

```python
# variable concnetración URM 
# Example of the Shapiro-Wilk Normality Test
data = totales_finales['concentracion_Nitrogen']
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probably Gaussian')
else:
 print('Probably not Gaussian')
```

```python
# variable concentración URM 
# Example of the Shapiro-Wilk Normality Test
data = totales_finales['Valor mg NO3/L']
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probably Gaussian')
else:
 print('Probably not Gaussian')
```

Información.
Un punto de decisión importante cuando se trabaja con una muestra de datos es si se deben utilizar métodos estadísticos paramétricos o no paramétricos.
Los métodos estadísticos paramétricos asumen que los datos tienen una distribución conocida y específica, a menudo una distribución gaussiana. Si una muestra de datos no es gaussiana, entonces se violan los supuestos de las pruebas estadísticas paramétricas, y se deben utilizar métodos estadísticos no paramétricos.
Hay una serie de técnicas que puede utilizar para comprobar si la muestra de datos se desvía de una distribución gaussiana, llamadas pruebas de normalidad.

1) Métodos gráficos para calificar las desviaciones de la normal, como los histogramas y el gráfico Q-Q.
2) Pruebas de normalidad estadística para cuantificar las desviaciones de lo normal.

La prueba de Shapiro-Wilk prueba la hipótesis nula de que los datos se extrajeron de una distribución normal. Éste nos dice que para p-valor > a 0.05, los datos provienen de una distribución normal.

Para las tres variables, se aplica el test de Shapiro-Wilk. 
En nuestro caso, las tres variables analizadas no siguen una distribución normal o gaussiana. A partir d'aquí, el análisis se enfoca a variables no paramétricas.

### 2. Prueba de Spearman’s Rank Correlation

Para el cálculo de la correlación, utilizamos el test de Spearman, ya que es adecuado para analizar variables no gaussianas o no parámetricas, como es nuestro caso.
La correlación de Spearman es adecuada para muestras con pocos valores, y variables no gaussianas. Los resultados son más robustos y precisos para este tipo de variables.
Se utiliza para demostrar si las variables son dependientes o independientes. Veremos si hay más o menos influencia entre las variables.
Se calcula numéricamente la correlación entre las variables de dos en dos, y esto se corroborará con los gráficos que se realizan más adelante.

```python
from scipy.stats import spearmanr
```

Correlación de Spearman entre  concentración de nitratos por deyección y concentracion URM.

```python
data1 = totales_finales['concentracion_URM']
data2 = totales_finales['concentracion_Nitrogen']
stat, p = spearmanr(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probably independent')
else:
 print('Probably dependent')
```

Correlación de spearman entre concentración de nitratos por deyección y Valor mg NO3/L.

```python
stat, p = spearmanr(totales_finales['Valor mg NO3/L'], totales_finales['concentracion_Nitrogen'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probably independent')
else:
 print('Probably dependent')
```

TEORIA
Error: Ignorar o no tener en cuenta las variables de confusión puede llevar a resultados de correlación engañosos. Las variables de confusión son factores externos que afectan a las dos variables estudiadas, haciendo que parezca que existe una correlación cuando en realidad no la hay, o al revés.

Ejemplo: Supongamos que estás analizando la relación entre el número de aplicaciones de protección solar y la incidencia de quemaduras solares. Encuentras una correlación negativa, lo que sugiere que más protección solar provoca más quemaduras solares. Sin embargo, la variable de confusión es el tiempo pasado al sol, que afecta tanto a la aplicación de protección solar como al riesgo de quemaduras solares.

En nuestro caso, es importante considerar la superficie de la comarca y utilizar las concentraciones en lugar de valores absolutos. Por otro lado, no hay que olvidar que la carga ganadera URM no es el único factor que influye en la contaminación de las aguas subterráneas.
