Primero voy a eliminar los nitrits. Selecciono los registros que son Nitratos

```python
seleccio = nitrats[nitrats['Variable'] == 'Nitrats']
```

```python
seleccio['Valor mg NO3/L'] = seleccio['Valor mg NO3/L'].str.replace('<', '')
```

Me voy a quedar con los registros que tengan un Valor de mg NO3/l superior a 50 mg/l, que es el limite de contaminación. Superior a estos valores son aguas contaminadas por nitratos, no aptas para el consumo humano.

```python
seleccio['Valor mg NO3/L'] = seleccio['Valor mg NO3/L'].astype(float)
```

APLICO EL FILTRO
Selecciono los valores que si representan aguas contaminadas, donde el límite esta en aquellos resultados superiores a 50 mg/l

```python
nitrats_final = seleccio[seleccio['Valor mg NO3/L'] >= 50]
```

```python
nitrats_final
```

Cambio los valores de coordenadas UTM a su valor correcto.

```python
nitrats_final['UTM X'] = nitrats_final['UTM X']*10
nitrats_final['UTM Y'] = nitrats_final['UTM Y']*10
```

Compruebo que el cambio de las coordenadas UTM se ha realizado correctamente

```python
nitrats_final
```

Listo los nombres de las masas de agua para ver como están escritos, y poder corregirlo

Creo una lista con todas las masas de agua y las comarcas a las que pertenece. Estos datos los hemos buscado previamente. La finalidad es tener unificarlo con los nitratos de las aguas para tener los datos por comarca.

```python
data = [
    ['Plana de Vic - Collsacabra', 'Osona'],
    ['Prelitoral del Vallès', 'Vallès Oriental'],
    ['Al·luvials del Vallès', 'Vallès Occidental'],
    ['Detrític neogen del Vallès', 'Vallès Occidental'],
    ['Plioquaternari de l\'Onyar', 'Gironès'],
    ['La Cerdanya', 'Cerdanya'],
    ['Prades - Alt Francolí', 'Conca de Barberà'],
    ['Detrític neogen de l\'Empordà', 'Alt Empordà'],
    ['Fluviodeltaic del Baix Ter', 'Baix Empordà'],
    ['Baixa Garrotxa-Pla de l\'Estany', 'Garrotxa'],
    ['Paleògens del Baix Ter', 'Baix Empordà'],
    ['Al·luvials de la Baixa Costa Brava', 'Baix Empordà'],
    ['Calcàries de Tàrrega', 'Urgell'],
    ['Al·luvial d\'Urgell', 'Urgell'],
    ['Al·luvials de la baixa i delta de la Tordera', 'Maresme'],
    ['Maresme', 'Maresme'],
    ['Baix Gaià', 'Tarragonès'],
    ['Garraf', 'Garraf'],
    ['Sorres de Santa Oliva', 'Baix Penedès'],
    ['Detrític neogen del Penedès', 'Alt Penedès'],
    ['Bloc del Gaià-Bonastre', 'Tarragonès'],
    ['Carme-Capellades', 'Anoia'],
    ['Montseny-Guilleries', 'Selva'],
    ['Moianès-Sant Llorenç del Munt', 'Moianès'],
    ['Serres marginals catalanes', 'Segarra'],
    ['Fluviovolcànic de la Garrotxa', 'Garrotxa'],
    ['Al·luvial del baix Segre_Cat', 'Segrià'],
    ['Tremp-Isona', 'Pallars Jussà'],
    ['Vall baixa i Delta del Llobregat', 'Baix Llobregat'],
    ['Alta Anoia', 'Anoia'],
    ['Calcàries de l\'Alt Foix-Gaià', 'Alt Camp'],
    ['Al·luvials del Ter Mig, Brugent i Llèmena', 'Gironès']
]
```

Creamos un DataFrame

```python
df_comarcas_aguas_contaminadas= pd.DataFrame(data, columns=['Lugar', 'Comarca'])
```

```python
df_comarcas_aguas_contaminadas
```

Modifico las palabras que no estan correctamente escritas en el archivo de las analíticas "nitrats_final"

```python
nitrats_final["massa d'aigua"] = nitrats_final["massa d'aigua"].replace('Vall�s', 'Vallès', regex=True)
nitrats_final["massa d'aigua"] = nitrats_final["massa d'aigua"].replace('Al�luvials', 'Al·luvials', regex=True)
nitrats_final["massa d'aigua"] = nitrats_final["massa d'aigua"].replace('Calc�ries', 'Calcàries', regex=True)
nitrats_final["massa d'aigua"] = nitrats_final["massa d'aigua"].replace('Detr�tic', 'Detrític', regex=True)
nitrats_final["massa d'aigua"] = nitrats_final["massa d'aigua"].replace('Francol�', 'Francolí', regex=True)
nitrats_final["massa d'aigua"] = nitrats_final["massa d'aigua"].replace('Empord�', 'Empordà', regex=True)
nitrats_final["massa d'aigua"] = nitrats_final["massa d'aigua"].replace('Alﾷluvials', 'Al·luvials', regex=True)
nitrats_final["massa d'aigua"] = nitrats_final["massa d'aigua"].replace('Pale�gens', 'Paleògens', regex=True)
nitrats_final["massa d'aigua"] = nitrats_final["massa d'aigua"].replace('Gai�', 'Gaià', regex=True)
nitrats_final["massa d'aigua"] = nitrats_final["massa d'aigua"].replace('Pened�s', 'Penedès', regex=True)
nitrats_final["massa d'aigua"] = nitrats_final["massa d'aigua"].replace('T�rrega', 'Tàrrega', regex=True)
nitrats_final["massa d'aigua"] = nitrats_final["massa d'aigua"].replace('Alﾷluvial', 'Al·luvial', regex=True)
nitrats_final["massa d'aigua"] = nitrats_final["massa d'aigua"].replace('Moian�s', 'Moianès', regex=True)
nitrats_final["massa d'aigua"] = nitrats_final["massa d'aigua"].replace('Lloren�', 'Llorenç', regex=True)
nitrats_final["massa d'aigua"] = nitrats_final["massa d'aigua"].replace('Fluviovolc�nic', 'Fluviovolcànic', regex=True)
nitrats_final["massa d'aigua"] = nitrats_final["massa d'aigua"].replace('Ll�mena', 'Llèmena', regex=True)
```

Comprueba que se han corrijido todos los errores de texto por carácteres especiales, y este el resultado

```python
nitrats_final["massa d'aigua"].unique()
```

Vamos a unir las tablas de nitrats_final y df_comarcas_aguas_contaminadas. Para añadir las comarcas a nuestra
tabla de resultados, como hemos comentado antes.

```python
df = pd.merge(nitrats_final, df_comarcas_aguas_contaminadas, how='inner', left_on='massa d\'aigua', right_on='Lugar')
```

```python
df
```

Calculo la media de los NO3 agrupados por data, lugar y comarca

```python
df_nitrats = df.groupby(['data', 'Lugar', 'Comarca'])['Valor mg NO3/L'].mean().round(2)
df_nitrats_comarca = pd.DataFrame(df_nitrats, columns=['Valor mg NO3/L']).reset_index()
df_nitrats_comarca
```

```python
df_nitrats_comarca['Comarca'].unique()
```

Paso a minúsculas los nombres de las comarcas, para posteriormente poder unirlo a los demás DataFrames

```python
df_nitrats_comarca['Comarca'] = df_nitrats_comarca['Comarca'].str.lower()
```

Previo a este paso, he hecho las uniones de tablas, y he visto que hay valores nulos porque la comarca no tiene el mismo nombre en las dos tablas. Vuelo a este punto a modificar el nombre de la comarca.

```python
df_nitrats_comarca['Comarca'] = df_nitrats_comarca['Comarca'].replace('selva', 'la selva')
df_nitrats_comarca
```

Me quedo con las medias de NO3 por comarca.

```python
nitrats_media_comarca = df_nitrats_comarca.groupby('Comarca')['Valor mg NO3/L'].mean().reset_index().round(2)
```
