# CALCULOS ESTADISTICOS PRIMERA APROXIMACION

### 1. Prueba de Shapiro-Wilk


Comprueba si una muestra de datos tiene una distribución gaussiana.
Suposiciones: Las observaciones en cada muestra son independientes y se distribuyen de manera idéntica (iid).
Interpretación:
H0: la muestra tiene una distribución gaussiana.
H1: la muestra no tiene una distribución gaussiana.

```python
from scipy.stats import shapiro
```

```python
# variable concnetración URM 
# Example of the Shapiro-Wilk Normality Test
data = totales_finales['concentracion_URM']
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probably Gaussian')
else:
 print('Probably not Gaussian')
```

```python
# variable concnetración URM 
# Example of the Shapiro-Wilk Normality Test
data = totales_finales['concentracion_Nitrogen']
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probably Gaussian')
else:
 print('Probably not Gaussian')
```

```python
# variable concentración URM 
# Example of the Shapiro-Wilk Normality Test
data = totales_finales['Valor mg NO3/L']
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probably Gaussian')
else:
 print('Probably not Gaussian')
```

Un punto de decisión importante cuando se trabaja con una muestra de datos es si se deben utilizar métodos estadísticos paramétricos o no paramétricos.
Un punto de decisión importante cuando se trabaja con una muestra de datos es si se deben utilizar métodos estadísticos paramétricos o no paramétricos.
Los métodos estadísticos paramétricos asumen que los datos tienen una distribución conocida y específica, a menudo una distribución gaussiana. Si una muestra de datos no es gaussiana, entonces se violan los supuestos de las pruebas estadísticas paramétricas y se deben utilizar métodos estadísticos no paramétricos.
Hay una serie de técnicas que puede utilizar para comprobar si la muestra de datos se desvía de una distribución gaussiana, llamadas pruebas de normalidad.

El hecho de que una muestra sea normal determina los tipos de métodos estadísticos que se deben utilizar con una muestra de datos.
Métodos gráficos para calificar las desviaciones de la normal, como los histogramas y el gráfico Q-Q.
Pruebas de normalidad estadística para cuantificar las desviaciones de lo normal.

La prueba de Shapiro-Wilk prueba la hipótesis nula de que el Los datos se extrajeron de una distribución normal.

Para las tres variables, se aplica el test de shapiro. Para p-valor > a 0.05, los datos estraidos provienen de una distribución normal.
En nuestro caso, las tres variables analizadas no siguen una distribución normal o gausiana. A partir d'aquí, el analisis se enfoca a varibales no paramétricas.

```python
import pandas as pd
from scipy.stats import mannwhitneyu

data = total_finales_sort4[['concentracion_Nitrogen','Valor mg NO3/L']]

df = pd.DataFrame(data)

# Seleccionar las dos columnas que quieres comparar
columna1 = df['concentracion_Nitrogen']
columna2 = df['Valor mg NO3/L']

# Realizar la prueba de Mann-Whitney U
stat, p_value = mannwhitneyu(columna1, columna2)
```

```python
print(f'Estadístico U: {stat}')
print(f'p-valor: {p_value}')
```

#### TEORIA
Interpretación de los resultados:
El estadístico U indica la magnitud de la diferencia entre los grupos.
El p-valor te dice si la diferencia entre los dos grupos es estadísticamente significativa.
Si el p-valor es menor que 0.05 (o tu nivel de significancia), puedes rechazar la hipótesis nula, lo que indicaría que las distribuciones de los dos grupos son diferentes.
Si el p-valor es mayor, no hay suficiente evidencia para rechazar la hipótesis nula.

Opciones adicionales:
Si quieres realizar la prueba como si fuera una prueba bilateral (dos colas) o unilateral (una cola), puedes usar el argumento alternative en mannwhitneyu:

 - alternative='two-sided': Prueba si las distribuciones son diferentes (por defecto).
 - alternative='less': Prueba si la distribución de la primera columna es menor que la de la segunda.
 - alternative='greater': Prueba si la distribución de la primera columna es mayor que la de la segunda.
